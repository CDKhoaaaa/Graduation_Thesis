```python
# GET COLUMN NAMES
find_heading_element = driver.find_element(By.XPATH, '//*[@id="contnet"]/div/div[2]/div/div/div[2]/div[1]/table/thead')
heading_elements = find_heading_element.find_elements(By.TAG_NAME, 'th')
table_headers = []
# Add the table header to the list
for heading_element in heading_elements:
    table_headers.append(heading_element.text)

table_headers_index = [i.replace('\n', " ").strip() for i in table_headers]
print(table_headers_index)
```

```python
# Get data rows
tbody_ = driver.find_element(By.XPATH, '//*[@id="contnet"]/div/div[2]/div/div/div[2]/div[1]/table/tbody')
td = tbody_.find_elements(By.TAG_NAME, 'td')
data = [] # Data rows
phase = []
for i in range(len(td)):
    # Remove 
    if td[i].text.startswith('Đợt'):
        phase.append(td[i])
        continue
    else:
        data.append(td[i].text)

print(data)
```

```python
# transform list to list of list
list_of_list_data = [data[i:i+6] for i in range(0, len(data), 6)]
list_of_list_data
```
