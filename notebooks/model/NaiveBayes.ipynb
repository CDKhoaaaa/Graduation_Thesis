```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.dummy import DummyClassifier # mô hình cơ sỏ
```

```python
import sys
import os
import pandas as pd
import numpy as np
```

```python
sys.path.append(os.path.abspath(os.path.join('..', 'utils')))

# Lấy thư mục hiện tại
current_directory = os.getcwd()
folder_path1 = os.path.join(current_directory,'../../data/train_test')
```

```python
x = pd.read_csv(f'{folder_path1}/xtrain_data.csv')
y = pd.read_csv(f'{folder_path1}/xtest_data.csv')
z = pd.read_csv(f'{folder_path1}/ytrain_data.csv')
t = pd.read_csv(f'{folder_path1}/ytest_data.csv')

x_train = x.iloc[:, 1:].values
x_test = y.iloc[:, 1:].values

y_train = z['drop_out'].values
y_test = t['drop_out'].values
```

```python
nb_model = GaussianNB()
```

```python
# Chuyển đổi y_train thành nhãn phân loại.
y_train_class = np.where(y_train >= 0.5, 1, 0)
y_test_class = np.where(y_test >= 0.5, 1, 0)
```

```python
nb_model.fit(x_train, y_train_class)
```

```python
y_pred_nb = nb_model.predict(x_test)
```

```python
y_pred_nb
```

```python
accuracy = accuracy_score(y_test_class, y_pred_nb)
conf_matrix = confusion_matrix(y_test_class, y_pred_nb)
class_report = classification_report(y_test_class, y_pred_nb)

print(f'Accuracy: {accuracy}')
print('Confusion Matrix:')
print(conf_matrix)
print('Classification Report:')
print(class_report)
```

```python
dummy_clf = DummyClassifier(strategy="most_frequent")
dummy_clf.fit(x_train, y_train)
baseline_prediction = dummy_clf.predict(x_test)
baseline_accuracy = accuracy_score(y_test_class, np.where(baseline_prediction==np.unique(baseline_prediction)[0],0,baseline_prediction))
```

```python
print("Model accuracy:", accuracy)
print("Baseline accuracy:", baseline_accuracy)
```

```python
# [[48 15]
#  [11 52]]


'''
TP: 48, TN: 52, FP: 15, FN: 11
'''
```

```python
x_test.shape[0]
```

```python
# Số lượng các mẫu trong tập dữ liệu kiểm tra (test set)
N = x_test.shape[0]

# Độ chính xác toàn thể (Overall Accuracy): (TN + TP) / N
overall_accuracy = round((48 + 52) / N * 100, 2)
print("Độ chính xác toàn thể: ", overall_accuracy)

# Sai số toàn thể (Overall Mistake): (FP + FN) / N
overall_mistake = round((15 + 11) / N * 100, 2)
print("Sai số toàn thể: ", overall_mistake)

# Độ nhạy (Sensitivity): TP / (TP + FN)
sensitivity = round((48 / (48 + 11)) * 100, 2)
print("Độ nhạy: ", sensitivity)

# Sai số của dự đoán sinh viên thôi học (Mistake Rate for Dropout Prediction)
mistake = round(100 - sensitivity, 2)
print("Tỉ lệ mô hình dự đoán sai sinh viên nghỉ học: ", mistake)

# Độ đặc hiệu (Specificity): TN / (TN + FP)
specificity = round((52 / (52 + 15)) * 100, 2)
print("Độ đặc hiệu: ", specificity)

# Sai số tỉ lệ mô hình dự đoán chính xác sinh viên không thôi học (Mistake Rate for Retention Prediction)
mistake_specificity = round(100 - specificity, 2)
print("Sai số tỉ lệ mô hình dự đoán chính xác sinh viên không thôi học: ", mistake_specificity)
```
